```python
#Saisir l'URL
url="https://en.wikipedia.org/wiki/Linkin_Park_discography"
```

```python
#importer le code de la page
#La fonction "session()" aide à l'objection ,nous permet de persister certains paramètres à travers les demandes
s= requests.Session()
reponse=s.get(url)
reponse
```

D'apres la ligne de code ci-dessus,nous avons obtenu une réponse [200], cela signifie essentiellement que le site a répondu avec un code de réponse HTTP 200 OK et nous a envoyé le contenu HTML de la page.



```python
contenu_p = reponse.text
print(contenu_p[:1000])
```

```python
#Enregistrons le contenu dans un fichier html

with open('projet.html', 'w', encoding="utf-8") as file:
    file.write(contenu_p)
```

```python
rows5 = required_table5.find_all("tr")
```

```python
#attributs d'en-tête du tableau
header5 = [th.text.rstrip() for th in rows5[0].find_all('th')]
header5 = header5[0:2] + [th.text.rstrip().split('[')[0] for th in rows5[1].find_all('th')] + header5[3:]
print(header5)
```

# Extraire les données

```python
#Extraire les données et ajoutez-les aux listes respectives
lst_data5 = []
c5= [ [], [], [], [], [], [], [], [], [], [], [], [], [], [] ]
for row in rows5[2:-1]:
    c5[0].append(row.find('th').text.strip())
    tds = row.find_all('td')
    if len(tds) == 13:
        for i, x in enumerate(tds):
            c5[i+1].append(x.text.strip())
    elif len(tds) == 12:
        if len(tds[0].text.strip()) == 4:
          for i, x in enumerate(tds):
              c5[i+1].append(x.text.strip())
          c5[13].append(c5[13][-1])
        else:
          c5[1].append(c5[1][-1])
          for i, x in enumerate(tds):
            c5[i+2].append(x.text.strip())
    else:
        c5[1].append(c5[1][-1])
        for i, x in enumerate(tds):
            c5[i+2].append(x.text.strip())
        c5[13].append(c5[13][-1])
```

```python
# créer un dictionnaire
d5 = dict()
for i, h in enumerate(header5):
  d5[h] = c5[i]
```

```python
# convertir dict en DataFrame
df_table5=pd.DataFrame(d5)
df_table5
```
